```python
import sys
import findspark
findspark.init()
from bs4 import BeautifulSoup
import pyspark
import lxml.html as html
import requests
import time
import sqlite3
```

```python
start_url = "http://www.vybory.izbirkom.ru/region/region/izbirkom?action=show&root=1&tvd=100100084849066&vrn=100100084849062&region=0&global=1&sub_region=0&prver=0&pronetvd=null&vibid=100100084849066&type=227"
response = requests.get(start_url)
page = html.document_fromstring(response.text)
tags = page.getchildren().pop()
regions = tags.xpath('//nobr//a')
```

```python
region_num = 1
for region in regions:
    region_name = region.text
    region_url = region.attrib['href']
    print(region_num, ' ', region_name)
    response = requests.get(region_url)
    region_page = html.document_fromstring(response.text)
    region_tags = region_page.getchildren().pop()
    tiks = region_tags.xpath('//nobr//a')
    
    for tik in tiks:
        tik_name = tik.text
        tik_url = tik.attrib['href']
        print('---', tik_name)
        
        response = requests.get(tik_url)
        response.headers
    print()
    region_num += 1
```

```python
conn = sqlite3.connect('task2.db')
c = conn.cursor()
c.execute('DROP TABLE IF EXISTS DATA')
c.execute('CREATE TABLE DATA(REGION TEXT DEFAULT NULL, TIK TEXT DEFAULT NULL, UIK TEXT DEFAULT NULL,'
          'VOTERS INT DEFAULT 0, BULLETINS INT DEFAULT 0, ANTICIPATORY INT DEFAULT 0, INSIDE INT DEFAULT 0, '
          'OUTSIDE INT DEFAULT 0, CANCEL INT DEFAULT 0, PORTABLE INT DEFAULT 0, STATIONARY INT DEFAULT 0, '
          'INVALID INT DEFAULT 0, VALID INT DEFAULT 0, LOST INT DEFAULT 0, UNACCOUNTED INT DEFAULT 0, '
          'BABURIN INT DEFAULT 0, GRUDININ INT DEFAULT 0, JIRINOVSKIY INT DEFAULT 0, PUTIN INT DEFAULT 0, '
          'SOBCHAK INT DEFAULT 0, SURAYKIN INT DEFAULT 0, TITON INT DEFAULT 0, YAVLINSKIY INT DEFAULT 0);')
start_time = time.time()
region_time = time.time()
for region in regions:
    region_name = region.text
    region_url = region.attrib['href']
    response = requests.get(region_url)
    region_page = html.document_fromstring(response.text)
    region_tags = region_page.getchildren().pop()
    tiks = region_tags.xpath('//nobr//a')
    
    for tik in tiks:
        tik_name = tik.text
        tik_url = tik.attrib['href']
        response = requests.get(tik_url)
        bs = BeautifulSoup(response.text, 'html.parser')
        links = bs.find_all('a', href=True)
        uik_url = links[5]['href']
        response = requests.get(uik_url).text
        soup = BeautifulSoup(response, 'html.parser')
        table = soup.find('table', attrs={'border': '0', 'bgcolor': '#ffffff', 'cellpadding': '2', 'cellspacing': '1',
                                       'style': 'width:100%;overflow:scroll', 'vspace': '0', 'align': 'left'})
        params_table = []
        for row in table.findAll('tr'):
            cells = row.findAll('td')
            params_table.append([cell.find(text=True) for cell in cells])
        
        for i in range(len(params_table[0])):
            query = f'INSERT INTO DATA VALUES ("{region_name}", "{tik_name}", "{params_table[0][i]}"'
            for j in range(1, 22):
                if j == 13:
                    continue
                query += f', {params_table[j][i]}'
            query += ');'
            c.execute(query)
        
        conn.commit()
    print(f'region {region_name} ready, time {time.time() - region_time}')
    region_time = time.time()

c.close()
conn.close()
print(f'full dbfill operation took {time.time() - start_time}')

#         uik_page = html.document_fromstring(response.text)
#         uik_tags = region_page.getchildren().pop()
#         uik_table = uik_tags.xpath('//td//table').pop()
        
#         uik_html_table = pd.read_html(html.tostring(uik_table), header = 0).pop()
```
